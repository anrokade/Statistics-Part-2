### 1. What is hypothesis testing in statistics?

Hypothesis testing is a fundamental concept in statistics used to assess whether there is enough evidence to support a particular claim or hypothesis about a population. It involves making inferences about population parameters based on sample data.

Here’s a simple breakdown of how it works:

1. State the Hypotheses
Null Hypothesis (H₀): This is the hypothesis that there is no effect or difference, and it represents the status quo. It's what you're trying to test against.

Alternative Hypothesis (H₁ or Ha): This is the hypothesis that there is an effect or difference. It's what you want to prove.

Example:

H₀: The mean test score of the population is 50.

H₁: The mean test score of the population is not 50.

2. Choose the Significance Level (α)
The significance level (usually denoted as α) is the probability of rejecting the null hypothesis when it is actually true. Common values for α are 0.05, 0.01, and 0.10.

3. Select the Appropriate Test
Depending on the data and the hypothesis, you select an appropriate statistical test (e.g., t-test, chi-square test, ANOVA, etc.). For example, if you're comparing means between two groups, you might use a t-test.

4. Calculate the Test Statistic
Using sample data, you calculate a test statistic (like t-value, z-value, or F-value) based on the type of test you're conducting.

5. Make a Decision
p-value: The p-value tells you how likely it is to observe the test results under the null hypothesis. If the p-value is less than or equal to α, you reject the null hypothesis (meaning the evidence supports the alternative hypothesis).

Decision rule: If the p-value is smaller than your significance level (α), reject H₀. If it's larger, fail to reject H₀.

6. Conclusion
If you reject the null hypothesis, you conclude that there is sufficient evidence to support the alternative hypothesis.

If you fail to reject the null hypothesis, you conclude that there is insufficient evidence to support the alternative hypothesis.

Example Scenario:
Suppose a school claims that the average height of its students is 150 cm. A researcher takes a sample of 30 students and finds an average height of 152 cm. To test this claim:

H₀: The average height is 150 cm (no difference).

H₁: The average height is not 150 cm (there is a difference).

Using a t-test, the researcher calculates the p-value. If the p-value is below a chosen threshold (e.g., 0.05), they reject H₀ and conclude that the average height is likely not 150 cm.

### 2. What is the null hypothesis, and how does it differ from the alternative hypothesis

The null hypothesis (H₀) and alternative hypothesis (H₁ or Ha) are two key components in hypothesis testing, and they represent opposing claims about a population or a data set.

1. Null Hypothesis (H₀):

- Definition: The null hypothesis is a statement that there is no effect, no difference, or no relationship between variables. It represents the status quo or a baseline assumption that researchers try to test against.

- Goal: In hypothesis testing, you assume the null hypothesis is true until evidence suggests otherwise. Essentially, you're testing whether the data provide enough evidence to reject this assumption.

- Nature: It is often a statement of "no change," "no difference," or "no effect."

Example of H₀:

- The mean weight of apples in a farm is 150 grams. (No difference from 150 grams.)

- The average test score in a class is 75. (No difference from 75.)

- There is no relationship between age and income in a population.

In hypothesis testing, rejecting H₀ means that the evidence points to an effect or a difference that contradicts the null hypothesis.

2. Alternative Hypothesis (H₁ or Ha):

- Definition: The alternative hypothesis is a statement that contradicts the null hypothesis. It asserts that there is an effect, difference, or relationship between the variables under investigation.

- Goal: The alternative hypothesis is what researchers are typically trying to prove or find evidence for. It represents the theory or claim you're testing.

- Nature: It is often a statement of "change," "difference," or "effect."

Example of H₁:

- The mean weight of apples in a farm is not 150 grams. (There is a difference.)

- The average test score in a class is not 75. (There is a difference.)

- There is a relationship between age and income in a population.

If you reject the null hypothesis (H₀), then the alternative hypothesis (H₁) becomes a plausible explanation of the data.

Key Differences:

| **Null Hypothesis (H₀)**                                                    | **Alternative Hypothesis (H₁ or Ha)**                                 |
| --------------------------------------------------------------------------- | --------------------------------------------------------------------- |
| Represents the **status quo** or **no effect** scenario.                    | Represents the **new claim** or the **effect** you're trying to find. |
| Assumes no difference or effect.                                            | Claims there is a difference or effect.                               |
| It is what you **test against** and try to **disprove**.                    | It is what you **hope to prove** or **support** with evidence.        |
| You fail to reject it if the data does not show strong evidence against it. | It becomes your conclusion if you reject the null hypothesis.         |


Example:

Imagine you are testing if a new drug is effective in lowering blood pressure. Here’s how the hypotheses might look:

- Null Hypothesis (H₀): The new drug has no effect on blood pressure. (This would mean the drug works no better than a placebo or has no effect at all.)

- Alternative Hypothesis (H₁): The new drug does have an effect on blood pressure. (This could mean it lowers blood pressure more than a placebo or has some other noticeable effect.)

How They Work Together:

- You gather data and perform a statistical test to determine if there’s enough evidence to reject H₀.

- If the test shows that the data is inconsistent with H₀, then you reject the null hypothesis in favor of H₁.

- If the data supports H₀, then you fail to reject the null hypothesis (which does not necessarily prove H₀ is true, just that there isn’t enough evidence against it).

In essence:

- H₀ is a starting point that you test and attempt to disprove.

- H₁ is what you hope to prove if the evidence supports it.

### 3. What is the significance level in hypothesis testing, and why is it important?

The significance level (denoted as α) is a critical concept in hypothesis testing. It defines the threshold at which you are willing to reject the null hypothesis (H₀) in favor of the alternative hypothesis (H₁). It essentially sets the standard for determining whether the results of a test are "statistically significant."

What is the Significance Level (α)?

- Definition: The significance level is the probability of rejecting the null hypothesis when it is actually true. In other words, it’s the likelihood of making a Type I error (false positive), where you incorrectly conclude that there is an effect or difference when, in fact, there isn't.

- Common Values: The most common significance levels are:

- α = 0.05 (5%): This means you're willing to accept a 5% chance of incorrectly rejecting the null hypothesis.

- α = 0.01 (1%): This means you're willing to accept only a 1% chance of a Type I error, which is a stricter standard.

- α = 0.10 (10%): Sometimes used when the consequences of a Type I error are less severe, or when you're testing a hypothesis with less certainty.

How it Works:

- In hypothesis testing, after you collect your data and perform the test, you compare the p-value (the probability of obtaining results at least as extreme as the observed results under the null hypothesis) with the significance level (α).

- If p-value ≤ α, you reject the null hypothesis (indicating there is sufficient evidence to support the alternative hypothesis).

- If p-value > α, you fail to reject the null hypothesis (indicating there is not enough evidence to support the alternative hypothesis).


Why is the Significance Level Important?

1) Defines the Risk of Type I Error:

- The significance level helps define the risk you’re willing to take of making a Type I error. For example, with α = 0.05, you accept that there’s a 5% chance you might incorrectly reject a true null hypothesis (i.e., a false positive).

2) Controls the Trade-off Between Type I and Type II Errors:

- The significance level is a trade-off between Type I error (false positive) and Type II error (false negative). A lower α (e.g., 0.01) reduces the risk of a Type I error but increases the risk of a Type II error (i.e., failing to reject a false null hypothesis).

- A higher α (e.g., 0.10) makes it easier to reject the null hypothesis, but it increases the risk of incorrectly rejecting it when it’s actually true.

3) Guides Decision Making:

- The significance level helps make clear-cut decisions in hypothesis testing. You either reject the null hypothesis or fail to reject it, based on whether your p-value meets the significance threshold.

4) Consistency Across Studies:

- By setting a standard significance level (like 0.05), researchers can communicate their results more consistently and allow for comparisons across different studies. If everyone uses the same α value, it provides a common metric for evaluating statistical evidence.

Example:

Let’s say you're testing whether a new drug lowers blood pressure more than a placebo, and you're using a significance level of α = 0.05.

1. You perform the statistical test and get a p-value = 0.03.

2. Since p-value ≤ α (0.05), you reject the null hypothesis. This suggests there is enough evidence to conclude that the drug has a significant effect on lowering blood pressure.

However, if your p-value = 0.08:

- Since p-value > α (0.05), you fail to reject the null hypothesis. This means that the evidence isn’t strong enough to claim the drug has a statistically significant effect.

What Happens if You Choose a Different α?

- α = 0.01 (1%): You need stronger evidence to reject the null hypothesis. This reduces the likelihood of a Type I error but increases the risk of a Type II error (missing a true effect).

- α = 0.10 (10%): You’re more likely to reject the null hypothesis, but this increases the chance of a Type I error. It may be used in exploratory or preliminary research, where you're willing to accept more risk.

Visualizing the Significance Level:

- Imagine you conduct a test and draw a normal distribution curve (assuming the null hypothesis is true). The area under the curve represents all possible outcomes of the test.

- The critical region (which is the region where you would reject H₀) is defined by the significance level, and it’s typically located in the tails of the distribution. For α = 0.05, you would have 2.5% in each tail of a two-tailed test.

### 7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

# ✅ 1. One-Tailed Test (Directional Test)

**Purpose**: Tests for the possibility of an effect in only one direction.

**Alternative Hypothesis (H₁)**: Specifies a direction — either greater than or less than.

**Example:**

- H₀: μ = 100  
- H₁: μ > 100  ← *Right-tailed*  
- or  
- H₁: μ < 100  ← *Left-tailed*

**Used when**: You're only interested in deviations in one direction.

🔍 **Example**:  
A company claims its battery lasts at least 10 hours. You test if it lasts less than 10 hours:

- H₀: μ ≥ 10  
- H₁: μ < 10  ← *One-tailed (left)*

---

# ✅ 2. Two-Tailed Test (Non-directional Test)

**Purpose**: Tests for the possibility of an effect in both directions.

**Alternative Hypothesis (H₁)**: Does *not* specify a direction.

**Example:**

- H₀: μ = 100  
- H₁: μ ≠ 100  ← *Two-tailed*

**Used when**: You're interested in any significant difference (either higher or lower).

🔍 **Example**:  
You want to test if a new teaching method has a different effect (better or worse) on test scores:

- H₀: No difference  
- H₁: There is a difference → *Two-tailed*

🎯 Key Differences:

| Feature         | One-Tailed Test                            | Two-Tailed Test               |
| --------------- | ------------------------------------------ | ----------------------------- |
| Direction       | One direction (>, <)                       | Both directions (≠)           |
| Critical Region | Entire $\alpha$ in one tail                | $\alpha/2$ in each tail       |
| Sensitivity     | More sensitive to effects in one direction | Less sensitive overall        |
| When to Use     | You have a specific directional hypothesis | You’re testing for any change |


### 8. What is the Z-test, and when is it used in hypothesis testing?

# 📊 What is the Z-Test?

The **Z-test** is a **statistical hypothesis test** used to determine whether there is a significant difference between a sample and population mean, or between two sample means, **when the population standard deviation is known** and the sample size is large (typically n>= 30 )).

---

## ✅ When to Use a Z-Test

Use a Z-test when **all of the following are true**:

1. The population standard deviation (σ) is **known**
2. The sample size is **large** (typically \( 𝑛
n≥30 Central Limit Theorem applies)
3. The data is **approximately normally distributed**
4. You're comparing:
   - A sample mean to a population mean
   - Two sample means
   - A sample proportion to a population proportion

---

## 🧮 Common Types of Z-Tests

| Z-Test Type             | Purpose                                          |
|-------------------------|--------------------------------------------------|
| One-sample Z-test       | Compare sample mean to known population mean     |
| Two-sample Z-test       | Compare means of two independent samples         |
| Z-test for proportions  | Compare sample proportion to population proportion or between two proportions |

---

## 🔍 Example: One-Sample Z-Test

A factory claims its light bulbs last an average of 1000 hours.  
You test 50 bulbs and find they last an average of 980 hours.  
The population standard deviation is known: \( σ = 80 \)

**Hypotheses:**

- H₀: μ = 1000  
- H₁: μ ≠ 1000

## 🧮 Z-Statistic Formula

$Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}$

**Where:**

* $\bar{x}$ = sample mean
* $\mu$ = population mean
* $\sigma$ = population standard deviation
* $n$ = sample size

## 📌 Summary

| Feature                 | Z-Test                        |
|-------------------------|-------------------------------|
| Known standard deviation (σ) | ✅ Required              |
| Sample size             | ✅ Typically large (≥ 30)     |
| Data distribution       | ✅ Normal or approx. normal   |
| Common use cases        | ✅ Mean and ✅ Proportion tests |


### 9.  How do you calculate the Z-score, and what does it represent in hypothesis testing?

## What Is a Z-Score?

A Z-score (or standard score) tells you how many standard deviations a data point or sample statistic is from the population mean. In hypothesis testing, the Z-score is used to determine how unusual or extreme a sample result is under the assumption that the null hypothesis is true.

## 🧮 Z-Score Formula

For a sample mean:

$Z = \frac{\bar{x} - \mu}{\sigma / \sqrt{n}}$

**Where:**

* $\bar{x}$ = sample mean
* $\mu$ = population mean under the null hypothesis
* $\sigma$ = population standard deviation
* $n$ = sample size

## 📌 What Does the Z-Score Represent?

The Z-score tells you how far your sample mean is from the population mean in units of standard error. It helps you determine how likely it is to observe your sample result if the null hypothesis is true.

## 🔍 Example Interpretation:

| Z-Score Value | Interpretation                                             |
| :------------ | :--------------------------------------------------------- |
| $Z = 0$       | Sample mean equals population mean                         |
| $Z = 2$       | Sample mean is 2 standard errors above the population mean |
| $Z = -1.5$    | Sample mean is 1.5 standard errors below the population mean |

## 🧠 In Hypothesis Testing:

1.  Calculate the Z-score using the sample data.
2.  Compare it to the critical Z-value based on your significance level (e.g., $\alpha = 0.05$).

**Decision Rule:**

* If the absolute Z-score is greater than the critical value $\rightarrow$ Reject $H_0$
* If not $\rightarrow$ Fail to reject $H_0$

## 🧪 Example:

Suppose:

* $\mu = 100$
* $\sigma = 15$
* $n = 36$
* $\bar{x} = 106$

Then:

$Z = \frac{106 - 100}{15 / \sqrt{36}} = \frac{6}{15 / 6} = \frac{6}{2.5} = 2.4$

If $\alpha = 0.05$ and you're doing a two-tailed test, the critical Z-value $\approx \pm 1.96$.

Since $2.4 > 1.96$, you reject the null hypothesis.

### 10. What is the T-distribution, and when should it be used instead of the normal distribution?

✅ What Is the T-Distribution?

The T-distribution (also called Student’s t-distribution) is a probability distribution that is similar to the normal distribution but has heavier tails. It accounts for extra uncertainty when estimating population parameters from small samples.

🧮 Key Characteristics of the T-Distribution:

* Symmetrical and bell-shaped, like the normal distribution.
* Wider tails — allows for more variability.
* Defined by degrees of freedom (df), usually $df = n - 1$.
* As $n \to \infty$, the t-distribution approaches the normal distribution.

🔁 When to Use the T-Distribution Instead of the Normal Distribution:

| Condition                       | Use T-Distribution? |
| :------------------------------ | :------------------ |
| Population standard deviation known | ❌ Use Z-distribution |
| Population standard deviation unknown | ✅ Use T-distribution |
| Sample size small (typically $n < 30$) | ✅ Use T-distribution |
| Sample size large and $\sigma$ unknown | ✅ T-distribution (but normal approximation is fine) |

🧠 Why Use the T-Distribution?

When you're estimating a mean and:

* You don’t know the population standard deviation ($\sigma$), and
* You have a small sample size,

you use the sample standard deviation ($s$) to estimate it — which introduces extra variability. The t-distribution adjusts for this.

🧮 T-Statistic Formula:

$t = \frac{\bar{x} - \mu}{s / \sqrt{n}}$

**Where:**

* $\bar{x}$ = sample mean
* $\mu$ = hypothesized population mean
* $s$ = sample standard deviation
* $n$ = sample size
* $df = n - 1$

🔍 Example:

You have a sample of 10 students, with:

* $\bar{x} = 72$
* $s = 8$
* testing against $\mu = 75$

$t = \frac{72 - 75}{8/\sqrt{10}} = \frac{-3}{8/3.162} = \frac{-3}{2.53} \approx -1.19$

Use a t-table with $df = 9$ to determine significance based on your $\alpha$ level.

🧾 Summary:

| Feature                       | T-Distribution        | Z (Normal) Distribution |
| ----------------------------- | --------------------- | ----------------------- |
| Use when σ is **unknown**     | ✅ Yes                 | ❌ No                    |
| Small sample size             | ✅ Yes                 | ❌ Not suitable          |
| Tails                         | Heavier (more spread) | Thinner (less spread)   |
| Depends on degrees of freedom | ✅ Yes                 | ❌ No                    |

### 12. What is the T-test, and how is it used in hypothesis testing?

✅ What is a T-Test?

A T-test is a statistical method used to determine whether there is a significant difference between:

* A sample mean and a known or hypothesized population mean, or
* The means of two groups

It’s used when the population standard deviation ($\sigma$) is unknown and the sample size is small (typically $n < 30$).

🧠 Why Use a T-Test?

The T-test accounts for extra uncertainty when estimating the population standard deviation from the sample. This is especially important with small samples, where the estimate is less reliable.

🧮 T-Test Formula (One-Sample):

$t = \frac{\bar{x} - \mu}{s / \sqrt{n}}$

**Where:**

* $\bar{x}$ = sample mean
* $\mu$ = population mean under $H_0$
* $s$ = sample standard deviation
* $n$ = sample size
* $df = n - 1$ = degrees of freedom

🧪 Types of T-Tests

| T-Test Type        | Used For                                                    |
| :----------------- | :---------------------------------------------------------- |
| One-sample t-test  | Comparing a sample mean to a known population mean          |
| Two-sample t-test  | Comparing means of two independent groups                   |
| Paired t-test      | Comparing means from the same group at two different times  |

📌 Steps to Use the T-Test in Hypothesis Testing

1.  **State the hypotheses**
    * $H_0$: No difference (e.g., $\mu = \mu_0$)
    * $H_1$: There is a difference (e.g., $\mu \neq \mu_0$)
2.  **Choose significance level** (commonly $\alpha = 0.05$)
3.  **Compute the t-statistic** using sample data
4.  **Find critical t-value** from t-distribution table (based on $df = n - 1$)
5.  **Compare the t-statistic with the critical value**
    * If $|t|$ > critical value $\rightarrow$ Reject $H_0$
    * Else $\rightarrow$ Fail to reject $H_0$

🔍 Example:

A teacher claims students score 70 on average. A sample of 10 students scores:

* $\bar{x} = 74$
* $s = 5$
* $n = 10$

$t = \frac{74 - 70}{5 / \sqrt{10}} = \frac{4}{5 / 3.162} = \frac{4}{1.58} \approx 2.53$

Degrees of freedom: $df = 9$

If critical value at $\alpha = 0.05$ (two-tailed) $\approx 2.262$

Since $2.53 > 2.262$, we reject $H_0$ $\rightarrow$ significant difference.

### 14. What is a confidence interval, and how is it used to interpret statistical results?

✅ Definition:

A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (such as the mean or proportion) with a certain level of confidence. In simpler terms, a CI gives an estimated range believed to include the true value you’re trying to measure — with a given level of certainty (e.g., 95%).

🧮 Confidence Interval Formula (for Mean):

$$ \text{CI} = \bar{x} \pm Z^* \left( \frac{\sigma}{\sqrt{n}} \right) $$

* $\bar{x}$: sample mean
* $Z^*$: Z-score corresponding to the desired confidence level
* $\sigma$: population standard deviation
* $n$: sample size

If $\sigma$ is unknown and $n$ is small, use the T-distribution instead of Z.

🔢 Common Confidence Levels:

| Confidence Level | Z-Value (approx.) |
| :--------------- | :---------------- |
| 90%              | 1.645             |
| 95%              | 1.96              |
| 99%              | 2.576             |

📌 Example:

Suppose:

* Sample mean ($\bar{x}$) = 80
* $\sigma = 10$
* $n = 25$
* 95% confidence level

Then:

$$ \text{CI} = 80 \pm 1.96 \left( \frac{10}{\sqrt{25}} \right) = 80 \pm 1.96 \times 2 = 80 \pm 3.92 $$

Result: $(76.08, 83.92)$

This means we are 95% confident that the true population mean lies between 76.08 and 83.92.

🧠 How to Interpret a Confidence Interval:

**Correct interpretation (for a 95% CI):**

“We are 95% confident that the true population parameter lies within this interval.”

**Incorrect interpretation:**

“There is a 95% chance that the parameter lies in this interval.”

⚠️ That’s a common mistake — once the interval is calculated, the true parameter either is or is not inside it.

🧪 Why It’s Useful:

* CI gives you a range instead of a single number (point estimate), showing the precision of your estimate.
* Wider intervals $\rightarrow$ more uncertainty
* Narrower intervals $\rightarrow$ more precision

### 15. What is the margin of error, and how does it affect the confidence interval?

✅ Definition:

The Margin of Error (MoE) is the maximum expected difference between the sample statistic (e.g., mean or proportion) and the true population parameter. It reflects how much uncertainty is associated with your estimate.

🧮 Formula:

For a population mean with a known standard deviation:

$$ \text{Margin of Error} = Z^* \times \left( \frac{\sigma}{\sqrt{n}} \right) $$

**Where:**

* $Z^*$ = Z-score based on the desired confidence level (e.g., 1.96 for 95%)
* $\sigma$ = population standard deviation (or use sample standard deviation $s$ if $\sigma$ is unknown)
* $n$ = sample size

Use a t-score instead of a Z-score when $\sigma$ is unknown and sample size is small.

📐 Confidence Interval and Margin of Error:

A confidence interval (CI) is built using the margin of error:

$$ \text{CI} = \text{Point Estimate} \pm \text{Margin of Error} $$

So, if:

* Sample mean = 70
* Margin of error = 5

Then:

CI = (70 − 5, 70 + 5) = (65, 75)

🔁 What Affects the Margin of Error?

| Factor                | Effect on Margin of Error             |
| :-------------------- | :------------------------------------ |
| Confidence Level $\uparrow$ | MoE $\uparrow$ (interval gets wider)  |
| Sample Size $\uparrow$    | MoE $\downarrow$ (interval gets narrower) |
| Standard Deviation $\uparrow$ | MoE $\uparrow$ (more variability = more error) |

🔍 Example:

Suppose:

* Sample mean = 60
* $\sigma = 12$
* $n = 36$
* Confidence level = 95% $\rightarrow Z^* = 1.96$

Then:

$$ \text{MoE} = 1.96 \times \left( \frac{12}{\sqrt{36}} \right) = 1.96 \times 2 = 3.92 $$

CI = 60 $\pm$ 3.92 $\rightarrow$ (56.08, 63.92)

🧠 Summary:

* The margin of error quantifies the uncertainty in your estimate.
* A larger MoE = wider confidence interval = less precision
* A smaller MoE = narrower confidence interval = greater precision

### 16. How is Bayes' Theorem used in statistics, and what is its significance?

✅ What Is Bayes' Theorem?

Bayes' Theorem is a fundamental concept in probability and statistics that describes how to update the probability of a hypothesis when given new evidence.

📐 Formula:

$$ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} $$

**Where:**

* $P(A|B)$: Posterior probability — probability of event A given B
* $P(B|A)$: Likelihood — probability of B given A is true
* $P(A)$: Prior probability — initial belief about A before seeing B
* $P(B)$: Marginal probability of B (sum of all ways B can happen)

🧠 Why It Matters:

Bayes' Theorem is powerful because it allows you to:

* Update beliefs as new evidence becomes available
* Combine prior knowledge with new data
* Use conditional probability to make better-informed decisions

🧪 Use Cases in Statistics:

1.  **Medical Testing**
    * A: Patient has the disease
    * B: Test result is positive

    You use Bayes' Theorem to calculate: $P(\text{Disease} | \text{Positive Test})$

    It helps answer: “What’s the chance a person actually has the disease if they test positive?”

2.  **Spam Filtering**
    * A: Email is spam
    * B: Email contains words like “free”, “win”, etc.

    $P(\text{Spam} | \text{Words in Email})$

    Bayesian filters learn from data and improve over time.

3.  **Bayesian Inference**
    In statistics:

    * Prior = Initial belief about a parameter (e.g., mean)
    * Likelihood = How likely your data is given the parameter
    * Posterior = Updated belief after seeing the data

    $$ \text{Posterior} \propto \text{Likelihood} \times \text{Prior} $$

    Used in:

    * Machine learning
    * Parameter estimation
    * Decision analysis

🔁 Contrast with Frequentist Statistics

| Feature           | Frequentist Approach          | Bayesian Approach                           |
| :---------------- | :---------------------------- | :------------------------------------------ |
| Probability       | Long-run frequency            | Degree of belief                            |
| Prior Knowledge   | Ignored                       | Incorporated through priors                 |
| Output            | Point estimates, p-values     | Posterior distributions                     |
| Updating          | Static                        | Dynamic (updates with new evidence)         |

📌 Significance of Bayes' Theorem

* Gives a mathematical way to update beliefs
* Foundation of Bayesian statistics, used in modern fields like:
    * Artificial Intelligence
    * Genetics
    * Finance
    * Clinical trials

🧾 Summary

Bayes' Theorem bridges prior knowledge and new evidence, allowing for dynamic and probabilistic reasoning — making it essential in many modern statistical applications.

### 17. What is the Chi-square distribution, and when is it used?

✅ Definition:

The Chi-Square distribution ($\chi^2$) is a continuous probability distribution commonly used in hypothesis testing and inferential statistics, especially when dealing with categorical data. It arises when you sum the squares of independent standard normal variables.

🧮 Formula Basis:

If $Z_1, Z_2, \dots, Z_k$ are independent standard normal variables (mean = 0, std dev = 1), then:

$$ \chi^2 = Z_1^2 + Z_2^2 + \dots + Z_k^2 $$

This sum follows a Chi-square distribution with $k$ degrees of freedom (df).

📌 Key Characteristics:

| Property    | Description                                              |
| :---------- | :------------------------------------------------------- |
| **Shape** | Right-skewed; becomes more symmetric as df $\uparrow$    |
| **Range** | Only non-negative values ($\geq 0$)                     |
| **Parameter** | Degrees of freedom (df)                                  |
| **Mean** | $\mu = \text{df}$                                        |
| **Variance** | $\sigma^2 = 2 \cdot \text{df}$                           |

🔍 When Is the Chi-Square Distribution Used?

1.  **Chi-Square Test for Independence**
    Used to determine whether two categorical variables are related.
    Example: Is gender independent of voting preference?
    Test compares observed frequencies vs. expected frequencies in a contingency table.

2.  **Chi-Square Goodness-of-Fit Test**
    Used to test whether an observed frequency distribution matches an expected distribution.
    Example: Do dice rolls follow a uniform distribution?

3.  **Testing Variance**
    Used to test hypotheses about population variance:

    $$ \chi^2 = \frac{(n-1) s^2}{\sigma^2} $$

🧠 Interpretation:

* Large $\chi^2$ values $\rightarrow$ larger deviation between observed and expected data $\rightarrow$ may reject null hypothesis.
* Degrees of freedom determine the shape of the distribution and critical values.

🧾 Summary:

| Feature           | Chi-Square Distribution              |
| :---------------- | :----------------------------------- |
| **Use Case** | Categorical data, variance testing   |
| **Data Type** | Frequencies (counts)                 |
| **Test Types** | Independence, Goodness-of-Fit, Variance |
| **Requirements** | Observations must be independent     |
| **Value Range** | $\chi^2 \geq 0$ (non-negative)      |

### 18. What is the Chi-square goodness of fit test, and how is it applied?

✅ Definition:

The Chi-Square Goodness-of-Fit Test is a statistical test used to determine whether the observed frequency distribution of a categorical variable matches an expected distribution.

It helps answer:

❓ “Do my observed data fit what I expected?”

🧮 Test Statistic Formula:

$$ \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i} $$

**Where:**

* $O_i$ = Observed frequency for category $i$
* $E_i$ = Expected frequency for category $i$

Sum is taken over all categories

📌 Hypotheses:

* **Null hypothesis $H_0$**: Observed distribution matches expected distribution
* **Alternative hypothesis $H_1$**: Observed distribution does not match expected distribution

🔢 Steps to Apply the Test:

1.  **Define Expected Frequencies**
    Based on a known or theoretical distribution (e.g., uniform, historical data)
2.  **Collect Observed Frequencies**
    From your sample or experiment
3.  **Calculate the Chi-square Statistic**
    $$ \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i} $$
4.  **Determine Degrees of Freedom**
    $df = k - 1$
    Where $k$ = number of categories
5.  **Compare with Critical Value or p-value**
    Use a chi-square distribution table (or software) to determine significance

📊 Example:

A die is rolled 60 times. You observe the following outcomes:

| Face | Observed (O) | Expected (E = 60/6 = 10) |
| :--- | :----------- | :----------------------- |
| 1    | 8            | 10                       |
| 2    | 12           | 10                       |
| 3    | 9            | 10                       |
| 4    | 11           | 10                       |
| 5    | 10           | 10                       |
| 6    | 10           | 10                       |

$$ \chi^2 = \frac{(8-10)^2}{10} + \frac{(12-10)^2}{10} + \dots = 1.0 $$

Degrees of freedom = $6 - 1 = 5$

Compare $\chi^2 = 1.0$ with critical value from table (e.g., 11.07 at 0.05 level)

✅ Since $1.0 < 11.07$, fail to reject $H_0$ $\rightarrow$ die is fair

⚠️ Assumptions:

* Categories are mutually exclusive
* Expected frequency in each category should be at least 5
* Sample data is random and independent

🧾 Summary:

| Component       | Description                                    |
| :-------------- | :--------------------------------------------- |
| **Purpose** | Compare observed vs. expected frequency distribution |
| **Data type** | Categorical                                    |
| **Test statistic** | $\chi^2 = \sum \frac{(O_i - E_i)^2}{E_i}$      |
| **Degrees of freedom** | $df = k - 1$                                   |
| **Common use cases** | Dice rolls, color preferences, genetics, etc.  |

### 19.  What is the F-distribution, and when is it used in hypothesis testing?

📊 What is the F-distribution, and When Is It Used in Hypothesis Testing?

✅ Definition:

The F-distribution is a continuous probability distribution that arises frequently in statistics, especially in the context of comparing variances. It is the distribution of the ratio of two independent chi-square variables, each divided by their respective degrees of freedom.

🧮 Mathematical Definition:

If $X \sim \chi^2_{d_1}$ and $Y \sim \chi^2_{d_2}$ are two independent chi-square random variables with degrees of freedom $d_1$ and $d_2$, then the random variable

$$ F = \frac{(X/d_1)}{(Y/d_2)} $$

follows an F-distribution with degrees of freedom $d_1$ (numerator) and $d_2$ (denominator).

📌 Key Characteristics:

| Feature    | Description                                             |
| :--------- | :------------------------------------------------------ |
| **Shape** | Right-skewed, varies with degrees of freedom            |
| **Range** | $F \geq 0$                                              |
| **Parameters** | Two degrees of freedom: numerator $d_1$, denominator $d_2$ |
| **Mean** | $\frac{d_2}{d_2 - 2}$ for $d_2 > 2$                     |

🔍 When Is the F-distribution Used?

1.  **ANOVA (Analysis of Variance)**
    * To test if means across multiple groups are equal
    * Compares variance between groups to variance within groups
    * Null hypothesis: All group means are equal
    * Test statistic follows an F-distribution under $H_0$

2.  **Testing Equality of Two Variances**
    * Comparing if two populations have the same variance
    * Null hypothesis: $\sigma_1^2 = \sigma_2^2$
    * Test statistic: ratio of sample variances, follows an F-distribution

3.  **Regression Analysis**
    * To test overall significance of a regression model
    * The F-test compares explained variance vs. unexplained variance

🧠 Interpretation:

* Large values of the F-statistic suggest greater variability between groups than within groups, leading to rejection of the null hypothesis.
* Critical values depend on the chosen significance level and degrees of freedom.

🧾 Summary:

| Aspect         | F-distribution                                      |
| :------------- | :-------------------------------------------------- |
| **Usage** | Comparing variances, ANOVA, regression              |
| **Data type** | Ratio of variances                                  |
| **Range** | $[0, \infty)$                                       |
| **Parameters** | Two degrees of freedom ($d_1, d_2$)                 |

### 20. What is an ANOVA test, and what are its assumptions?

✅ What is ANOVA?

ANOVA (Analysis of Variance) is a statistical method used to compare the means of three or more groups to determine if at least one group mean is significantly different from the others.

It tests the null hypothesis that all group means are equal:

$H_0: \mu_1 = \mu_2 = \mu_3 = \dots = \mu_k$

Against the alternative hypothesis that at least one mean differs.

🧮 How ANOVA Works:

ANOVA compares between-group variability to within-group variability. It calculates an F-statistic:

$$ F = \frac{\text{Variance Between Groups}}{\text{Variance Within Groups}} $$

If $F$ is significantly large, it suggests group means are not all equal.

📋 Assumptions of ANOVA:

* **Independence of Observations:**
    Data points in each group must be independent of each other.
* **Normality:**
    The data in each group should be approximately normally distributed.
* **Homogeneity of Variances (Homoscedasticity):**
    The variance among the groups should be approximately equal.

🧠 Why Are These Assumptions Important?

* Violations can affect the validity of the ANOVA results.
* If assumptions are violated, alternatives like Welch’s ANOVA or non-parametric tests (e.g., Kruskal-Wallis) may be used.

🧾 Summary:

| Aspect         | Description                                        |
| :------------- | :------------------------------------------------- |
| **Purpose** | Compare means of 3+ groups                         |
| **Null Hypothesis** | All group means are equal                          |
| **Test Statistic** | F-ratio (Between-group variance / Within-group variance) |
| **Key Assumptions** | Independence, Normality, Equal variances           |

### 21. What are the different types of ANOVA tests?

📊 Different Types of ANOVA Tests

ANOVA tests come in several varieties, each suited for different experimental designs and data structures. Here are the most common types:

1.  **One-Way ANOVA**
    * **Purpose:** Compare means across one independent categorical variable (factor) with three or more groups.
    * **Example:** Comparing test scores of students from three different teaching methods.
    * **Assumptions:** Normality, homogeneity of variances, independence.

2.  **Two-Way ANOVA (Factorial ANOVA)**
    * **Purpose:** Compare means across two independent categorical variables and investigate:
        * Main effect of each factor
        * Interaction effect between the two factors
    * **Example:** Comparing effects of teaching method (factor 1) and gender (factor 2) on test scores.
    * **Assumptions:** Same as One-Way ANOVA.

3.  **Repeated Measures ANOVA**
    * **Purpose:** Used when the same subjects are measured under multiple conditions or times.
    * **Example:** Measuring blood pressure of patients before, during, and after treatment.
    * **Key Feature:** Accounts for correlation between repeated measurements.
    * **Assumptions:** Normality, sphericity (equal variances of differences).

4.  **Multivariate ANOVA (MANOVA)**
    * **Purpose:** Extension of ANOVA to multiple dependent variables simultaneously.
    * **Example:** Testing the effect of diet on both weight and cholesterol levels.
    * **Assumptions:** Multivariate normality, homogeneity of covariance matrices.

5.  **ANCOVA (Analysis of Covariance)**
    * **Purpose:** Like ANOVA but controls for one or more continuous covariates (confounding variables).
    * **Example:** Comparing group means while adjusting for age.
    * **Assumptions:** Same as ANOVA + linearity between covariates and dependent variable.

🧾 Summary Table

| Type                    | Number of Factors   | Repeated Measures? | Multiple DVs? | Controls Covariates? |
| :---------------------- | :------------------ | :----------------- | :------------ | :------------------- |
| **One-Way ANOVA** | 1                   | No                 | No            | No                   |
| **Two-Way ANOVA** | 2                   | No                 | No            | No                   |
| **Repeated Measures ANOVA** | 1+ (same subjects)  | Yes                | No            | No                   |
| **MANOVA** | 1+                  | No                 | Yes           | No                   |
| **ANCOVA** | 1+ (with covariates) | No                 | No            | Yes                  |

### 22. What is the F-test, and how does it relate to hypothesis testing?

✅ What is the F-test?

The F-test is a statistical test that compares variances to determine if they are significantly different from each other. It uses the F-distribution, which is the distribution of the ratio of two scaled chi-square distributions.

🧮 How It Works:

The F-test calculates the ratio:

$$ F = \frac{\text{Variance}_1}{\text{Variance}_2} $$

**Where:**

* $\text{Variance}_1$ = variance of the first sample/group
* $\text{Variance}_2$ = variance of the second sample/group

If the variances are equal, the ratio will be close to 1. A large ratio suggests a significant difference between variances.

📌 Hypothesis Setup:

* **Null hypothesis $H_0$**: The variances are equal
    $H_0: \sigma_1^2 = \sigma_2^2$
* **Alternative hypothesis $H_1$**: The variances are not equal
    $H_1: \sigma_1^2 \neq \sigma_2^2$

🔍 Relation to Hypothesis Testing:

* The F-test is often used before comparing means to check the assumption of equal variances (homogeneity of variance).
* It is the basis for ANOVA tests, which compare means across multiple groups by examining the ratio of between-group variance to within-group variance.
* A significant F-test (large F statistic) suggests rejecting the null hypothesis, indicating variances (or means in ANOVA context) differ.

🧾 Summary:

| Aspect        | Description                                   |
| :------------ | :-------------------------------------------- |
| **Purpose** | Compare variances of two or more groups       |
| **Test statistic** | Ratio of variances (F-ratio)                  |
| **Distribution** | F-distribution                                |
| **Common uses** | Testing equality of variances, ANOVA          |
| **Null hypothesis** | Variances are equal                           |

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pmf(x, p):
    """
    Calculate the probability of x for a Bernoulli distribution with parameter p.

    Parameters:
    x (int): The value (0 or 1)
    p (float): Probability of success (0 <= p <= 1)

    Returns:
    float: Probability P(X = x)
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return p**x * (1 - p)**(1 - x)

# Example usage:
p = 0.7  # Probability of success
print(f"P(X=1) = {bernoulli_pmf(1, p)}")
print(f"P(X=0) = {bernoulli_pmf(0, p)}")
```

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05, alternative='two-sided'):
    """
    Perform a Z-test for a sample mean against a population mean.

    Parameters:
    - sample: array-like, sample data
    - pop_mean: float, population mean under H0
    - pop_std: float, population standard deviation (assumed known)
    - alpha: significance level (default 0.05)
    - alternative: 'two-sided', 'greater', or 'less'

    Returns:
    - z_stat: computed Z statistic
    - p_value: p-value of the test
    - conclusion: whether to reject H0 or not
    """
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate standard error
    se = pop_std / np.sqrt(n)

    # Calculate Z statistic
    z_stat = (sample_mean - pop_mean) / se

    # Calculate p-value based on alternative hypothesis
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
    elif alternative == 'less':
        p_value = norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'greater', or 'less'")

    # Conclusion
    if p_value < alpha:
        conclusion = "Reject the null hypothesis"
    else:
        conclusion = "Fail to reject the null hypothesis"

    return z_stat, p_value, conclusion

# Example usage:
np.random.seed(42)
sample_data = np.random.normal(loc=102, scale=15, size=30)  # Sample data
population_mean = 100
population_std = 15
alpha = 0.05
alternative = 'two-sided'  # Change to 'greater' or 'less' as needed

z_stat, p_value, conclusion = z_test(sample_data, population_mean, population_std, alpha, alternative)
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")
print(f"Conclusion: {conclusion}")
```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95, pop_std=None):
    """
    Calculate the confidence interval for the mean.

    Parameters:
    - data: array-like, sample data
    - confidence: float, confidence level (default 0.95)
    - pop_std: float or None, population standard deviation. If None, uses sample std and t-distribution.

    Returns:
    - (lower_bound, upper_bound): confidence interval
    """
    n = len(data)
    mean = np.mean(data)

    if pop_std is not None:
        # Use Z-distribution
        se = pop_std / np.sqrt(n)
        z = norm.ppf((1 + confidence) / 2)
        margin_of_error = z * se
    else:
        # Use sample std and t-distribution
        from scipy.stats import t
        sample_std = np.std(data, ddof=1)
        se = sample_std / np.sqrt(n)
        t_val = t.ppf((1 + confidence) / 2, df=n-1)
        margin_of_error = t_val * se

    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=50)

# Population std unknown, use t-distribution
conf_level = 0.95
lower, upper = confidence_interval(sample_data, confidence=conf_level)

print(f"{int(conf_level*100)}% Confidence Interval for the mean: ({lower:.2f}, {upper:.2f})")

# Interpretation
print(f"This means we are {int(conf_level*100)}% confident that the true population mean lies between {lower:.2f} and {upper:.2f}.")
```

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    """
    Perform a one-sample Z-test for the mean.

    Parameters:
    - sample: array-like, sample data
    - population_mean: float, known population mean under H₀
    - population_std: float, known population standard deviation
    - alpha: float, significance level (default = 0.05)
    - alternative: str, 'two-sided', 'greater', or 'less'

    Returns:
    - z_stat: Z test statistic
    - p_value: p-value of the test
    - conclusion: interpretation based on the test
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    standard_error = population_std / np.sqrt(n)

    z_stat = (sample_mean - population_mean) / standard_error

    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
    elif alternative == 'less':
        p_value = norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'greater', or 'less'")

    # Interpretation
    if p_value < alpha:
        result = "Reject the null hypothesis (significant difference)."
    else:
        result = "Fail to reject the null hypothesis (no significant difference)."

    return z_stat, p_value, result

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=105, scale=10, size=30)  # Sample data
known_pop_mean = 100
known_pop_std = 10
alpha = 0.05
alternative = 'greater'  # Options: 'two-sided', 'greater', 'less'

z_stat, p_value, conclusion = z_test(sample_data, known_pop_mean, known_pop_std, alpha, alternative)

print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")
print(f"Conclusion: {conclusion}")
```

```python
# 3.  Implement a one-sample Z-test using Python to compare the sample mean with the population mean

import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, pop_mean, pop_std, alpha=0.05, alternative='two-sided'):
    """
    Perform a one-sample Z-test.

    Parameters:
    - sample: array-like, sample data
    - pop_mean: float, known population mean
    - pop_std: float, known population standard deviation
    - alpha: significance level (default = 0.05)
    - alternative: 'two-sided', 'greater', or 'less'

    Returns:
    - z_stat: computed Z-statistic
    - p_value: computed p-value
    - conclusion: decision on the null hypothesis
    """
    sample_mean = np.mean(sample)
    n = len(sample)
    se = pop_std / np.sqrt(n)

    z_stat = (sample_mean - pop_mean) / se

    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
    elif alternative == 'less':
        p_value = norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'greater', or 'less'")

    # Decision
    if p_value < alpha:
        conclusion = "Reject the null hypothesis"
    else:
        conclusion = "Fail to reject the null hypothesis"

    return z_stat, p_value, conclusion

# 🔢 Example Usage
np.random.seed(42)
sample_data = np.random.normal(loc=105, scale=10, size=50)  # Sample from N(105, 10)
known_pop_mean = 100
known_pop_std = 10
alpha = 0.05
alternative = 'greater'  # Try 'two-sided', 'greater', or 'less'

z_stat, p_val, result = one_sample_z_test(sample_data, known_pop_mean, known_pop_std, alpha, alternative)

print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_val:.4f}")
print(f"Conclusion: {result}")
```

```python
# 4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    se = pop_std / np.sqrt(n)
    z_stat = (sample_mean - pop_mean) / se
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    if p_value < alpha:
        conclusion = "Reject the null hypothesis"
    else:
        conclusion = "Fail to reject the null hypothesis"

    return z_stat, p_value, conclusion

# Simulate sample data
np.random.seed(1)
sample_data = np.random.normal(loc=102, scale=10, size=50)
pop_mean = 100
pop_std = 10
alpha = 0.05

# Perform Z-test
z_stat, p_val, decision = two_tailed_z_test(sample_data, pop_mean, pop_std, alpha)

# Print results
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_val:.4f}")
print(f"Conclusion: {decision}")

# Plot the decision regions
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

z_crit = norm.ppf(1 - alpha/2)

plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Standard Normal Distribution')

# Shade critical regions
plt.fill_between(x, y, where=(x <= -z_crit), color='red', alpha=0.3, label='Rejection Region (α/2)')
plt.fill_between(x, y, where=(x >= z_crit), color='red', alpha=0.3)

# Mark Z-statistic
plt.axvline(z_stat, color='blue', linestyle='--', label=f'Z-stat = {z_stat:.2f}')
plt.axvline(-z_crit, color='black', linestyle='--', label=f'Critical Value = ±{z_crit:.2f}')
plt.axvline(z_crit, color='black', linestyle='--')

plt.title('Two-Tailed Z-Test Decision Region')
plt.xlabel('Z-value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu_0, mu_1, sigma, n, alpha=0.05, alternative='greater'):
    """
    Visualize Type I and Type II errors for a Z-test.

    Parameters:
    - mu_0: mean under H0
    - mu_1: actual mean under H1
    - sigma: known population standard deviation
    - n: sample size
    - alpha: significance level
    - alternative: 'greater', 'less', or 'two-sided'
    """
    se = sigma / np.sqrt(n)

    # Critical value under H0
    if alternative == 'greater':
        z_crit = norm.ppf(1 - alpha)
        crit_value = mu_0 + z_crit * se
        beta = norm.cdf((crit_value - mu_1) / se)
    elif alternative == 'less':
        z_crit = norm.ppf(alpha)
        crit_value = mu_0 + z_crit * se
        beta = 1 - norm.cdf((crit_value - mu_1) / se)
    elif alternative == 'two-sided':
        z_crit = norm.ppf(1 - alpha/2)
        crit_upper = mu_0 + z_crit * se
        crit_lower = mu_0 - z_crit * se
        beta = (norm.cdf((crit_upper - mu_1) / se) -
                norm.cdf((crit_lower - mu_1) / se))
    else:
        raise ValueError("alternative must be 'greater', 'less', or 'two-sided'")

    # X values for plot
    x = np.linspace(mu_0 - 4*se, mu_1 + 4*se, 1000)
    h0_dist = norm.pdf(x, mu_0, se)
    h1_dist = norm.pdf(x, mu_1, se)

    plt.figure(figsize=(10, 6))

    # Plot distributions
    plt.plot(x, h0_dist, label='H₀: μ = {:.1f}'.format(mu_0), color='blue')
    plt.plot(x, h1_dist, label='H₁: μ = {:.1f}'.format(mu_1), color='green')

    # Shade Type I Error (α)
    if alternative == 'greater':
        plt.fill_between(x, 0, h0_dist, where=(x >= crit_value), color='red', alpha=0.4, label='Type I Error (α)')
        plt.fill_between(x, 0, h1_dist, where=(x < crit_value), color='orange', alpha=0.3, label='Type II Error (β)')
        plt.axvline(crit_value, color='black', linestyle='--', label='Critical Value')
    elif alternative == 'less':
        plt.fill_between(x, 0, h0_dist, where=(x <= crit_value), color='red', alpha=0.4, label='Type I Error (α)')
        plt.fill_between(x, 0, h1_dist, where=(x > crit_value), color='orange', alpha=0.3, label='Type II Error (β)')
        plt.axvline(crit_value, color='black', linestyle='--', label='Critical Value')
    elif alternative == 'two-sided':
        plt.fill_between(x, 0, h0_dist, where=(x >= crit_upper) | (x <= crit_lower), color='red', alpha=0.4, label='Type I Error (α)')
        plt.fill_between(x, 0, h1_dist, where=(x > crit_lower) & (x < crit_upper), color='orange', alpha=0.3, label='Type II Error (β)')
        plt.axvline(crit_upper, color='black', linestyle='--', label='Critical Values')
        plt.axvline(crit_lower, color='black', linestyle='--')

    # Plot settings
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print β and power
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

visualize_type1_type2_errors(mu_0=100, mu_1=105, sigma=15, n=30, alpha=0.05, alternative='greater')
```

```python
# 6.  Write a Python program to perform an independent T-test and interpret the results

import numpy as np
from scipy.stats import ttest_ind

# Step 1: Generate two independent sample datasets
np.random.seed(42)
group1 = np.random.normal(loc=100, scale=10, size=30)  # Sample 1: Mean = 100
group2 = np.random.normal(loc=105, scale=10, size=30)  # Sample 2: Mean = 105

# Step 2: Perform the independent t-test
t_stat, p_value = ttest_ind(group1, group2, equal_var=True)  # Assumes equal variance

# Step 3: Display results
print("📊 Independent T-Test Results")
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpret the result
alpha = 0.05
if p_value < alpha:
    print("✅ Conclusion: Reject the null hypothesis — significant difference between the groups.")
else:
    print("❌ Conclusion: Fail to reject the null hypothesis — no significant difference between the groups.")
```

```python
# 7.  Perform a paired sample T-test using Python and visualize the comparison results

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

# Step 1: Simulate paired data (e.g., before and after treatment)
np.random.seed(42)
before = np.random.normal(loc=100, scale=8, size=30)
after = before + np.random.normal(loc=-2, scale=5, size=30)  # slight decrease after treatment

# Step 2: Perform the paired t-test
t_stat, p_value = ttest_rel(before, after)

# Step 3: Display results
print("📊 Paired Sample T-Test Results")
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("✅ Conclusion: Reject the null hypothesis — significant difference between the paired groups.")
else:
    print("❌ Conclusion: Fail to reject the null hypothesis — no significant difference.")

# Step 4: Visualize the comparison
plt.figure(figsize=(10, 5))

# Plot individual connections
for i in range(len(before)):
    plt.plot([0, 1], [before[i], after[i]], color='gray', alpha=0.5)

# Plot means
plt.scatter(np.zeros_like(before), before, color='blue', label='Before')
plt.scatter(np.ones_like(after), after, color='orange', label='After')

plt.xticks([0, 1], ['Before', 'After'])
plt.ylabel('Score / Measurement')
plt.title('Paired Sample Comparison')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 8.  Simulate data and perform both Z-test and T-test, then compare the results using Python

import numpy as np
from scipy.stats import norm, ttest_1samp

def one_sample_z_test(sample, pop_mean, pop_std):
    n = len(sample)
    sample_mean = np.mean(sample)
    se = pop_std / np.sqrt(n)
    z_stat = (sample_mean - pop_mean) / se
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # two-tailed
    return z_stat, p_value

# Step 1: Simulate sample data
np.random.seed(42)
sample = np.random.normal(loc=102, scale=10, size=25)  # true mean = 102
pop_mean = 100
pop_std = 10  # known for Z-test only

# Step 2: Perform Z-test (assuming known population standard deviation)
z_stat, z_pval = one_sample_z_test(sample, pop_mean, pop_std)

# Step 3: Perform T-test (sample std used)
t_stat, t_pval = ttest_1samp(sample, pop_mean)

# Step 4: Display results
print("📊 Z-test vs T-test Comparison (Two-tailed)")
print(f"{'Test':<10} {'Stat':>10} {'P-value':>15}")
print("-" * 37)
print(f"{'Z-test':<10} {z_stat:>10.3f} {z_pval:>15.4f}")
print(f"{'T-test':<10} {t_stat:>10.3f} {t_pval:>15.4f}")

# Step 5: Interpretation
alpha = 0.05
print("\n📌 Conclusion:")
if z_pval < alpha:
    print("✅ Z-test: Reject the null hypothesis")
else:
    print("❌ Z-test: Fail to reject the null hypothesis")

if t_pval < alpha:
    print("✅ T-test: Reject the null hypothesis")
else:
    print("❌ T-test: Fail to reject the null hypothesis")
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
from scipy import stats

def confidence_interval_mean(sample, confidence=0.95):
    """
    Calculates the confidence interval for the sample mean.

    Parameters:
    - sample (array-like): The sample data.
    - confidence (float): Confidence level (default 0.95 for 95% CI).

    Returns:
    - (lower_bound, upper_bound): Tuple representing the confidence interval.
    """
    sample = np.array(sample)
    n = len(sample)
    mean = np.mean(sample)
    sem = stats.sem(sample)  # standard error of the mean
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)  # critical t-value

    margin_of_error = t_crit * sem
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    return lower, upper

# 🔢 Example usage
np.random.seed(42)
data = np.random.normal(loc=50, scale=8, size=30)
ci = confidence_interval_mean(data, confidence=0.95)

print(f"📦 95% Confidence Interval: ({ci[0]:.2f}, {ci[1]:.2f})")
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
from scipy import stats

def calculate_margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for a sample mean.

    Parameters:
    - sample (array-like): The sample data
    - confidence (float): The desired confidence level (e.g., 0.95 for 95%)

    Returns:
    - margin_of_error (float)
    """
    sample = np.array(sample)
    n = len(sample)
    sem = stats.sem(sample)  # standard error of the mean
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n - 1)  # critical t-value
    moe = t_crit * sem
    return moe

# 🔢 Example usage
np.random.seed(1)
sample_data = np.random.normal(loc=100, scale=12, size=40)
confidence_level = 0.95

margin_of_error = calculate_margin_of_error(sample_data, confidence=confidence_level)
print(f"📏 Margin of Error at {int(confidence_level * 100)}% confidence: ±{margin_of_error:.2f}")
```

```python
# 11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior, likelihood, false_positive_rate):
    """
    Calculate posterior probability using Bayes' Theorem.

    Parameters:
    - prior: P(H), prior probability of hypothesis (disease)
    - likelihood: P(E|H), probability of evidence given hypothesis (positive test if diseased)
    - false_positive_rate: P(E|not H), probability of evidence if hypothesis is false (positive test if healthy)

    Returns:
    - posterior: P(H|E), updated probability of hypothesis given evidence
    """
    # P(not H)
    prior_neg = 1 - prior

    # Total probability of evidence P(E)
    prob_evidence = likelihood * prior + false_positive_rate * prior_neg

    # Bayes' Theorem
    posterior = (likelihood * prior) / prob_evidence
    return posterior

# Given data
prior_disease = 0.01
likelihood_positive_if_disease = 0.99
false_positive_rate = 0.05

posterior_prob = bayes_theorem(prior_disease, likelihood_positive_if_disease, false_positive_rate)
print(f"Posterior Probability of Disease given Positive Test: {posterior_prob:.4f} ({posterior_prob*100:.2f}%)")
```

```python
# 12. Perform a Chi-square test for independence between two categorical variables in Python

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Sample data: Contingency table of two categorical variables
# Example: Smoking status (Smoker/Non-Smoker) vs Lung Disease (Yes/No)
data = np.array([[30, 10],   # Smoker: 30 with disease, 10 without
                 [20, 40]])  # Non-Smoker: 20 with disease, 40 without

# Convert to DataFrame (optional, for better readability)
df = pd.DataFrame(data, columns=['Disease', 'No Disease'], index=['Smoker', 'Non-Smoker'])
print("Contingency Table:\n", df)

# Perform Chi-square test
chi2_stat, p_val, dof, expected = chi2_contingency(data)

print(f"\nChi-square Statistic: {chi2_stat:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p_val:.4f}")
print("\nExpected Frequencies:\n", expected)

# Interpretation
alpha = 0.05
if p_val < alpha:
    print("\nReject the null hypothesis - variables are associated.")
else:
    print("\nFail to reject the null hypothesis - variables are independent.")
```

```python
# 13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test of independence.

    Parameters:
    - observed: 2D numpy array or list of lists representing observed frequencies.

    Returns:
    - expected: 2D numpy array of expected frequencies.
    """
    observed = np.array(observed)
    row_sums = observed.sum(axis=1).reshape(-1, 1)  # Sum of each row (column vector)
    col_sums = observed.sum(axis=0).reshape(1, -1)  # Sum of each column (row vector)
    total = observed.sum()

    expected = (row_sums @ col_sums) / total  # Outer product divided by total sum
    return expected

# Example observed data (contingency table)
observed_data = [
    [30, 10],   # Category 1
    [20, 40]    # Category 2
]

expected_freq = calculate_expected_frequencies(observed_data)

print("Observed Frequencies:")
print(np.array(observed_data))
print("\nExpected Frequencies:")
print(expected_freq)
```

```python
# 14.  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution

import numpy as np
from scipy.stats import chisquare

# Observed frequencies (example)
observed = np.array([50, 30, 20])

# Expected frequencies (example) — must sum to same total as observed
expected = np.array([40, 40, 20])

# Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject null hypothesis: observed data does NOT fit the expected distribution.")
else:
    print("Fail to reject null hypothesis: observed data fits the expected distribution.")
```
